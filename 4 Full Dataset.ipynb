```python
import pandas as pd
variables = pd.DataFrame.from_csv("Variables.csv", encoding="UTF-8")
variables.columns = ["Index","Oil","Gold","NaturalGas"]
```

```python
import csv
with open("stocksArray.csv", 'rt') as f:
    reader = csv.reader(f)
    for row in reader:
        stocksArray = row
```

```python
temp = []
for x in stocksArray:
    if x == "LMT":
        x = "NYSE:LMT"
    if x =="NWL":
        x = "NYSE:NWL"
    if x=="NBL":
        x = "NYSE:NBL"
    temp.append(x)
stocksArray = temp
```

```python
import pandas_datareader.data as web
import datetime
df = variables
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2017, 1, 1)
for x in stocksArray:
    stock = web.DataReader(x, 'google', start, end)
    stock = stock["Close"].pct_change()
    stock.name = x
    df = pd.concat([df,stock],axis=1)
    print(x)
print(df)
```

```python
df.to_csv("StockData.csv", encoding="UTF-8")
```
