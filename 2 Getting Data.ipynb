```python
import datetime
startDate = datetime.datetime(2010, 1, 1)
endDate = datetime.datetime(2017, 5, 1)
```

```python
import pandas_datareader.data as web
stock = web.DataReader("F", 'google', startDate, endDate)
```

```python
print(stock)
```

```python
print(stock["Close"])
```

```python
print(stock[["Close","Open"]])
```

```python
print(stock.pct_change())
```

```python
df = stock.pct_change()["Close"]
print(df)
```

```python
df.dropna(inplace=True)
```

```python
index = web.DataReader("SPY", 'google', startDate, endDate)
print(index)
```

```python
oil = web.DataReader("DCOILWTICO", "fred", startDate, endDate)
print(oil)
```

```python
gold = web.DataReader("GOLDAMGBD228NLBM", "fred", startDate, endDate)
print(gold)
```

```python
naturalGas = web.DataReader("DHHNGSP", "fred", startDate, endDate)
print(naturalGas)
```

```python
index = index["Close"].pct_change()
oil = oil.pct_change()
gold = gold.pct_change()
naturalGas = naturalGas.pct_change()
```

```python
import pandas as pd
df = pd.concat([index, oil,gold,naturalGas], axis=1)
print(df)
```

```python
df.dropna(inplace=True)
print(df)
```

```python
df.to_csv("Variables.csv", encoding="UTF-8")
```
